```python
import pandas as pd
```

```python
filename="XAUUSD1440"

df = pd.read_csv("D:\\LABS\\histdata\\xauusd\\"+filename+".csv", header=0, 
                 names=['Date', 'Time', 'Open','High','Low','Close','Volume'], 
                 parse_dates=[['Date', 'Time']])
df = df.set_index('Date_Time', drop=False)
lowest = df.groupby([pd.Grouper(freq='M')])[['Date_Time','Close']].min()
lowest.reset_index(drop=True, inplace=True)
#lowest.to_excel('LOWEST-'+filename+'.xls')
```

```python
import mplfinance as fplt
```

```python
mc = fplt.make_marketcolors(
                            up='tab:blue',down='tab:red',
                            edge='lime',
                            wick={'up':'blue','down':'red'},
                            volume='tab:green',
                           )

s  = fplt.make_mpf_style(marketcolors=mc)

fplt.plot(
            df,
            type='line',
            style=s,
            title='XAUUSD 1997 - 2021',
            ylabel='Price (USD)',
            volume=True,
            ylabel_lower='Volume\nTraded',
            show_nontrading=True
            )
```

```python
def dataframe_per_year(year)

    return df[df["Date_Time"].dt.year == year]
```
